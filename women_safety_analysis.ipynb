# **Import Required Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Load the Dataset**

```python
# Load dataset
df = pd.read_excel("women_safety_dataset.xlsx")

# Display first 5 rows
print("First 5 Rows of Dataset:")
print(df.head())
```

## **Basic Dataset Information**

```python
print("\nDataset Information:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())
```

# **Data Cleaning**

```python
# Fill missing values with 0 (or you can use mean/median)
df.fillna(0, inplace=True)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

print("\nData Cleaning Completed")
```

**Exploratory Data Analysis (EDA)**

```python
print("\nAREA Incidents:")
print(df['area'].value_counts())
print("\nCrime Type Count:")
print(df['crime_type'].value_counts())
```

Visualization â€“ Crime Types

```python
plt.figure(figsize=(8,5))
df['crime_type'].value_counts().plot(kind='bar')
plt.title("Types of Incidents Against Women")
plt.xlabel("Crime Type")
plt.ylabel("Number of Cases")
plt.xticks(rotation=45)
plt.show()
```

# **Visualization â€“ Location Risk Analysis**

```python
plt.figure(figsize=(8,5))
df['area'].value_counts().head(10).plot(kind='bar')
plt.title("Top Risk Locations")
plt.xlabel("Area")
plt.ylabel("Number of Incidents")
plt.xticks(rotation=45)
plt.show()
```

# **Visualization â€“ Incident Percentage (Pie Chart)**

```python
plt.figure(figsize=(6,6))
df['crime_type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Crime Type Distribution")
plt.ylabel("")
plt.show()
```

# **Simple Insights Generation**

```python
most_common_incident = df['crime_type'].value_counts().idxmax()
most_risky_location = df['area'].value_counts().idxmax()

print("\nKey Insights:")
print("Most Common Incident Type:", most_common_incident)
print("Most Risky Location:", most_risky_location)
```

# **Save Cleaned Data**

```python
df.to_excel("cleaned_women_safety_data.xlsx", index=False)
print("\nCleaned dataset saved successfully")
```
